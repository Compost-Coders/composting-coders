## 1. Compost vs. Ambient Temperature Analysis
This dataset tracks the internal core temperature of a compost pile against the external ambient temperature over a full yearly cycle.

* **Thermophilic Peak (May 2025):** The pile hit a maximum of **60°C** shortly after initiation, despite outside temperatures being only ~10°C. This confirms high microbial activity.
* **Thermal Resilience (Winter 2025/26):** During extreme cold snaps where the outside temperature dropped to **-30°C**, the compost core maintained a stable range of **15°C to 30°C**. 
* **Thermal Delta:** The system achieved a maximum temperature differential ($\Delta T$) of approximately $50^{\circ}\text{C}$, indicating excellent insulation properties of the organic mass.
* **Data Anomaly:** A sharp drop to $0^{\circ}\text{C}$ in late December suggests a sensor or power interruption rather than a biological cooling event.

## 2. Active Heating & Energy Correlation
This analysis compares the **Internal Temperature** (Red) with the **Active Heating Power** (Blue) required to sustain the pile through winter.

* **Self-Sustaining Phase (May – Oct):** Heating power remained at **0 Watts**. The heat during this period was 100% biological (thermogenesis).
* **Active Maintenance (Nov – March):** As ambient temperatures fell, external heating was introduced to prevent dormancy. Power consumption fluctuated between **30W and 60W**.
* **System Failure Correlation:** The simultaneous "zeroing" of both temperature and wattage in late December confirms an **external power failure**.
* **Energy Efficiency:** A relatively low peak input of **60 Watts** was sufficient to counteract a $-30^{\circ}\text{C}$ external environment, keeping the core at a "living" temperature of $\sim 20^{\circ}\text{C}$.
