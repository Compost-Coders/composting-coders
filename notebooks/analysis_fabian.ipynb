## 1. Compost vs. Ambient Temperature Analysis
This dataset tracks the internal core temperature of a compost pile against the external ambient temperature over a full yearly cycle.

* **Thermophilic Peak (May 2025):** The pile hit a maximum of **60°C** shortly after initiation, despite outside temperatures being only ~10°C. This confirms high microbial activity.
* **Thermal Resilience (Winter 2025/26):** During extreme cold snaps where the outside temperature dropped to **-30°C**, the compost core maintained a stable range of **15°C to 30°C**. 
* **Thermal Delta:** The system achieved a maximum temperature differential ($\Delta T$) of approximately $50^{\circ}\text{C}$, indicating excellent insulation properties of the organic mass.
* **Data Anomaly:** A sharp drop to $0^{\circ}\text{C}$ in late December suggests a sensor or power interruption rather than a biological cooling event.

## 2. Active Heating & Energy Correlation
This analysis compares the **Internal Temperature** (Red) with the **Active Heating Power** (Blue) required to sustain the pile through winter.

* **Self-Sustaining Phase (May – Oct):** Heating power remained at **0 Watts**. The heat during this period was 100% biological (thermogenesis).
* **Active Maintenance (Nov – March):** As ambient temperatures fell, external heating was introduced to prevent dormancy. Power consumption fluctuated between **30W and 60W**.
* **System Failure Correlation:** The simultaneous "zeroing" of both temperature and wattage in late December confirms an **external power failure**.
* **Energy Efficiency:** A relatively low peak input of **60 Watts** was sufficient to counteract a $-30^{\circ}\text{C}$ external environment, keeping the core at a "living" temperature of $\sim 20^{\circ}\text{C}$.

# Thermal Analysis: November Incident

## 1. Compost System (Heating & Temp)
The "November Fall" in the compost bin was a result of **external thermal displacement**.
* **Observation:** As `Outside-Outside-Temperature` dropped, `Compost 1-Inside-Heating - w` increased to its maximum (~50W).
* **The Gap:** The heating power was insufficient to maintain the target temperature against the cold ambient air. Internal temperatures fell because the heat loss through the bin walls exceeded the energy input from the heater.
* **Conclusion:** The compost bin requires better **insulation** to decouple internal temperatures from the outside environment.

## 2. Growing Tents (Sensors 01-12)
The sensors in the tents show a different pattern, focusing on **environmental consistency**.
* **Pre-Collapse:** All 12 sensors showed typical fluctuating moisture levels, indicating active irrigation or plant transpiration.
* **The Shutdown:** Around the same time as the compost temperature fall, all Growing sensors (01-12) dropped to zero.
* **Post-Collapse Recovery:** In March, only **Sensors 01-06** recovered. **Sensors 07-12** remain flat at zero.
* **Hardware Insight:** This suggests that the "November Fall" might have been a power or gateway event that affected the Growing Tent infrastructure specifically, with a permanent hardware/connection failure for the second half of the sensors (07-12).

| System | Primary Issue | Action |

```python
series = df_compost_pd['Compost 1-Middle-Temperature'].interpolate(method='linear')

# 2. Perform STL Decomposition (Period=7 for weekly or 1 for daily if data is daily)
stl = STL(series, period=7, robust=True)
result = stl.fit()

# 3. Plotting
fig, (ax1, ax2, ax3, ax4) = plt.subplots(4, 1, figsize=(12, 10), sharex=True)

result.observed.plot(ax=ax1, color='blue')
ax1.set_ylabel('Observed')
ax1.set_title('Compost Temperature: STL Decomposition')

result.trend.plot(ax=ax2, color='red')
ax2.set_ylabel('Trend')

result.seasonal.plot(ax=ax3, color='green')
ax3.set_ylabel('Seasonal')

result.resid.plot(ax=ax4, color='orange', style='o')
ax4.set_ylabel('Residuals')

plt.tight_layout()
plt.show()
```

The **Residuals** plot represents the data left over after removing the long-term trend and the daily seasonal cycles. It captures "shocks" to the system—events that cannot be explained by regular patterns.

### Key Observations:
* **The November Anomaly:** There is a cluster of negative residuals in mid-November. This proves that the temperature drop was not just a "cold day," but a structural collapse where the heating system failed to maintain its predicted capacity.
* **The Noise Ceiling:** Notice how the residuals are relatively stable (small "noise") during the summer months, but become extremely volatile once the external temperature drops. This indicates that the system becomes statistically **unstable** in cold conditions.
* **March 2026 Rebound:** The sharp positive spikes in the residuals at the end of the timeline represent the manual intervention/restart. These are "positive shocks" where the system was forced back into a higher temperature state.

### Conclusion for the Report:
The residuals act as a **System Health Monitor**. A healthy, well-insulated compost bin would show random, low-amplitude noise. The high-amplitude "spikes" seen here are the mathematical fingerprint of a system that is too vulnerable to its environment.
