## 1. Compost vs. Ambient Temperature Analysis
This dataset tracks the internal core temperature of a compost pile against the external ambient temperature over a full yearly cycle.

* **Thermophilic Peak (May 2025):** The pile hit a maximum of **60°C** shortly after initiation, despite outside temperatures being only ~10°C. This confirms high microbial activity.
* **Thermal Resilience (Winter 2025/26):** During extreme cold snaps where the outside temperature dropped to **-30°C**, the compost core maintained a stable range of **15°C to 30°C**. 
* **Thermal Delta:** The system achieved a maximum temperature differential ($\Delta T$) of approximately $50^{\circ}\text{C}$, indicating excellent insulation properties of the organic mass.
* **Data Anomaly:** A sharp drop to $0^{\circ}\text{C}$ in late December suggests a sensor or power interruption rather than a biological cooling event.

```python
fig, ax1 = plt.subplots(figsize=(12, 6))

# Temperature (Line)
ax1.set_xlabel('Day')
ax1.set_ylabel('Temperature (°C)', color="red", fontsize=12, fontweight='bold')
ax1.plot(df_compost_pd['Day'], df_compost_pd['Compost 1-Middle-Temperature'], color="tab:red", label='Temperature inside')
ax1.tick_params(axis='y', labelcolor="tab:red")
ax1.grid(True, alpha=0.3)

# Heating
ax2 = ax1.twinx() 
ax2.set_ylabel('Heating (Watts)', color="purple", fontsize=12, fontweight='bold')

# fill everything purple if heating exceeds temperature
ax2.fill_between(df_compost_pd['Day'], 
                 df_compost_pd['Compost 1-Middle-Temperature'], 
                 df_compost_pd["Compost 1-Inside-Heating - w"], 
                 where = df_compost_pd["Compost 1-Inside-Heating - w"] > df_compost_pd["Compost 1-Middle-Temperature"],
                 color="purple", alpha=0.3, label='Heating (Backup)')
                
# fill everything green if heating is less than temperature
ax2.fill_between(df_compost_pd['Day'], 
                 df_compost_pd['Compost 1-Middle-Temperature'], 
                 df_compost_pd["Compost 1-Inside-Heating - w"], 
                 where = df_compost_pd["Compost 1-Inside-Heating - w"] < df_compost_pd["Compost 1-Middle-Temperature"],
                 color="green", alpha=0.3, label='"Bio-heating"(Active)')

plt.title("Relation between Inside Temperature & Active Heating", fontsize=14)
fig.tight_layout()
plt.show()
```

## 2. Active Heating & Energy Correlation
This analysis compares the **Internal Temperature** (Red) with the **Active Heating Power** (Blue) required to sustain the pile through winter.

* **Self-Sustaining Phase (May – Oct):** Heating power remained at **0 Watts**. The heat during this period was 100% biological (thermogenesis).
* **Active Maintenance (Nov – March):** As ambient temperatures fell, external heating was introduced to prevent dormancy. Power consumption fluctuated between **30W and 60W**.
* **System Failure Correlation:** The simultaneous "zeroing" of both temperature and wattage in late December confirms an **external power failure**.
* **Energy Efficiency:** A relatively low peak input of **60 Watts** was sufficient to counteract a $-30^{\circ}\text{C}$ external environment, keeping the core at a "living" temperature of $\sim 20^{\circ}\text{C}$.


```python
# get data from the composts
compost_1_df = df_compost.select(pl.col("Day"),
                                  pl.col("Compost 1-Inside-Heating - kWh").alias("Inside Heating -kWh"),
                                  pl.col("Compost 1-Inside-Heating - w").alias("Inside Heating -w"),
                                  pl.col("Compost 1-Lower-Moisture").alias("Lower-Moisture"),
                                  pl.col("Compost 1-Lower-Temperature").alias("Lower-Temperature"),
                                  pl.col("Compost 1-Middle-Moisture").alias("Middle-Moisture"),
                                  pl.col("Compost 1-Middle-Temperature").alias("Middle-Temperature"),
                                  pl.col("Compost 1-Upper-Moisture").alias("Upper-Moisture"),
                                  pl.col("Compost 1-Upper-Temperature").alias("Upper-Temperature")
                                  )
# label the data
compost_1_df = compost_1_df.with_columns(Source=pl.lit("Compost 1"))

compost_2_df = df_compost.select(
                                  pl.col("Day"),
                                  pl.col("Compost 2-Inside-Heating - kWh").alias("Inside Heating -kWh"),
                                  pl.col("Compost 2-Inside-Heating - w").alias("Inside Heating -w"),
                                  pl.col("Compost 2-Lower-Moisture").alias("Lower-Moisture"),
                                  pl.col("Compost 2-Lower-Temperature").alias("Lower-Temperature"),
                                  pl.col("Compost 2-Middle-Moisture").alias("Middle-Moisture"),
                                  pl.col("Compost 2-Middle-Temperature").alias("Middle-Temperature"),
                                  pl.col("Compost 2-Upper-Moisture").alias("Upper-Moisture"),
                                  pl.col("Compost 2-Upper-Temperature").alias("Upper-Temperature")
                                  )      
# label the data      
compost_2_df = compost_2_df.with_columns(Source=pl.lit("Compost 2"))  

# concatenate
composts_df = pl.concat([compost_1_df, compost_2_df])


# unpivor(melt) the data
df_compost_long = composts_df.unpivot(
    index=["Day", "Source"], 
    variable_name="Metric", 
    value_name="Value"
)

# plot a relplot with composts as hue                                      
sns.relplot(
    data=df_compost_long, 
    x="Day", 
    y="Value", 
    col="Metric",
    hue="Source", 
    kind="line", 
    facet_kws={'sharey': False},
    col_wrap=3,
    height=4,
    aspect=2.5,
)
plt.show()
```

```python
sensor_df = df_compost.select(
                              pl.col("Day"),
                              pl.col("Growing-Sensor 01-Moisture").alias("Sensor 1"),
                              pl.col("Growing-Sensor 02-Moisture").alias("Sensor 2"),
                              pl.col("Growing-Sensor 03-Moisture").alias("Sensor 3"),
                              pl.col("Growing-Sensor 04-Moisture").alias("Sensor 4"),
                              pl.col("Growing-Sensor 05-Moisture").alias("Sensor 5"),
                              pl.col("Growing-Sensor 06-Moisture").alias("Sensor 6"),
                              pl.col("Growing-Sensor 07-Moisture").alias("Sensor 7"),
                              pl.col("Growing-Sensor 08-Moisture").alias("Sensor 8"),
                              pl.col("Growing-Sensor 09-Moisture").alias("Sensor 9"),
                              pl.col("Growing-Sensor 10-Moisture").alias("Sensor 10"),
                              pl.col("Growing-Sensor 11-Moisture").alias("Sensor 11"),
                              pl.col("Growing-Sensor 12-Moisture").alias("Sensor 12"))

zeros = sensor_df.select(pl.all() == 0).mean() * 100
zeros
```

We can observe that zeros are present in all sensors. They make up about 50 - 60% on average. Due to moisture of 0 being fairly unlikely I suppose that's because the sensors weren't working properly.

```python
fig, ax = plt.subplots(figsize=(12,8))
# use pandas for bisualization
mean_sensor = sensor_df.select(pl.all()).mean().to_pandas()

# plot the mean moisture level of each sensor
ax.set_title("Average Moisture Level of Sensors")
ax.set_xlabel("Sensor")
ax.set_ylabel("Moisture (%)")
plt.xticks(rotation=45)

# show the data
sns.barplot(mean_sensor)
fig.tight_layout()
plt.show()
```
